```python
#Vectores
''' 
Caracterización: diccionario
La definición estricta de un vector es la de una entidad matemática que tiene
tanto magnitud (o longitud) como dirección. Permite representarlo gráficamente 
y algrebáicamente.

Vectores fila y Vectores columna.
'''
alumnos_aidone = {
    'Rodrigo':[5,4],
    'Lucia':[2,-3],
    'Alejandro':[3,5],
    'Valeria':[-5,4],
    'Javier':[0,-1],
    'Camila': [3,2],
    'Diego':[-1,1],
    'Gabriela':[5,-2],
    'Mateo':[-5,3],
    'Sofia':[-5,1]
}

''' 
Representación de vectores
Distancia = distancia entre vertices de vectores. 
'''
```

```python
import math

def distancia_2d(vec1,vec2,precision = 2): #precision es num de decimales 
    suma = 0
    for indice_componente in range(2):
        suma += (vec1[indice_componente]-vec2[indice_componente])**2
    distancia = math.sqrt(suma)
    distancia = round(distancia,precision)
    return distancia

vec_lucia = alumnos_aidone['Lucia']
for alumno_ref in ['Rodrigo', 'Gabriela', 'Mateo']:
    vec_ref = alumnos_aidone[alumno_ref]
    distancia = distancia_2d(vec_lucia, vec_ref)
    print('Distancia Lucia -', alumno_ref,distancia)
```

## Producto escalar





```python
def dot_product(vec1,vec2,precision=2):
    producto = 0
    for indice in range(len(vec1)): 
        producto += vec1[indice]*vec2[indice]
    producto = round(producto,precision)
    return producto
```

## Normas, magnitudes y longitudes

Norma = |Vector|





```python
def norm_vec(vec,precision=1):
    escalar = dot_product(vec,vec,precision)
    resultado = math.sqrt(escalar)
    resultado = round(resultado,precision)
    return resultado

vect_lucia = alumnos_aidone['Lucia']
norm_vec(vec_lucia,1)
```

## Angulo entre dos vectores





```python
''' Construimos sobre funciones que ya tenemos. A mayor coseno, mayor similitud'''
def cos_vec(vec1,vec2,precision = 1):
    resultado = dot_product(vec1,vec2)/(norm_vec(vec1)*norm_vec(vec2))
    resultado = round(resultado,precision)
    return resultado
```
